# Fichier de téléchargement des données

```python
import os
import requests
import pandas as pd
import py7zr
fold_download = "../../RAW"
subfold = "LCZ"
fold_save = "../../PROD"
```

## Téléchargement des données LCZ

```python
# Première étape : Téléchargement de la liste des resources disponibles sur le site data.gouv.fr
# Lien de la page : https://www.data.gouv.fr/fr/datasets/5d13a8b6634f41070a43dff3/ ; copier ci-dessous le lien du dataset export-resource
api_ressources = "https://www.data.gouv.fr/fr/datasets/r/4babf5f2-6a9c-45b5-9144-ca5eae6a7a6d"
response = requests.get(api_ressources)
with open(os.path.join(fold_download,"ressources.csv"), 'wb') as f:
    f.write(response.content)
```

```python
# Deuxième étape : Téléchargement des données LCZ
api = pd.read_csv(os.path.join(fold_download,"ressources.csv"), sep=";")
to_download = api[(api['dataset.title']=="Cartographie des Zones Climatiques Locales (LCZ) de 83 aires urbaines de plus de 50 000 habitants - 2022") &(api['format']=="zip")].filter(items=['title','url'])
to_download.to_csv(os.path.join(fold_download,subfold,"to_download.csv"), index=False)

os.makedirs(os.path.join(fold_download,subfold), exist_ok=True)

# starting with st nazaire
# to_download = to_download[to_download['title'].str.contains("saint-nazaire")]

for i in range(len(to_download)):
    print("Downloading "+to_download.iloc[i]['title']+" "+ str(i+1)+"/"+str(len(to_download)))
    url = to_download.iloc[i]['url']
    response = requests.get(url)
    with open(os.path.join(os.path.join(fold_download,subfold),to_download.iloc[i]['title']), 'wb') as f:
        f.write(response.content)

    folder_files = to_download.iloc[i]['title'].replace('.zip',"")
    # unzip file
    os.system("unzip -o "+os.path.join(os.path.join(fold_download,subfold),to_download.iloc[i]['title'])+" -d "+os.path.join(fold_download,subfold,folder_files))
    # remove zip file
    os.system("rm "+os.path.join(os.path.join(fold_download,subfold),to_download.iloc[i]['title']))
    # remove only .pdf, .qml and .tiff files
    os.system("rm "+os.path.join(fold_download,subfold,folder_files,"*.pdf"))
    os.system("rm "+os.path.join(fold_download,subfold,folder_files,"*.tif"))
    os.system("rm "+os.path.join(fold_download,subfold,folder_files,"*.qml"))
```

## Téléchargement des données FILOSOFI / carrés de 200m

```python
# Première étape : téléchargement des données sur le site de l'INSEE
# Dernières données dispo : 2019
# Lien de la page : https://www.insee.fr/fr/statistiques/7655475?sommaire=7655515
filo_carreau_Link = "https://www.insee.fr/fr/statistiques/fichier/7655475/Filosofi2019_carreaux_200m_shp.zip"
subfold = "INSEE/FILOSOFI/2019/CARREAUX_200M"
os.makedirs(os.path.join(fold_download,subfold), exist_ok=True)

response = requests.get(filo_carreau_Link)


with open(os.path.join(fold_download,subfold,"Filosofi2019_carreaux_200m_shp.zip"), 'wb') as f:
    f.write(response.content)

# unzip file
os.system("unzip -o "+os.path.join(fold_download,subfold,"Filosofi2019_carreaux_200m_shp.zip")+" -d "+os.path.join(fold_download,subfold))
with py7zr.SevenZipFile(os.path.join(fold_download,subfold,"Filosofi2019_carreaux_200m_shp.7z"), 'r') as archive:
    archive.extractall(path=os.path.join(fold_download,subfold))

# remove zip and 7z file
os.system("rm "+os.path.join(fold_download,subfold,"Filosofi2019_carreaux_200m_shp.zip"))
os.system("rm "+os.path.join(fold_download,subfold,"Filosofi2019_carreaux_200m_shp.7z"))
```
