```python
import geopandas as gpd
import os
folder_raw = "/Users/yann/Documents/Meio/R&D MILOO/RAW/ADMIN-EXPRESS-COG-CARTO_3-2__SHP_LAMB93_FXX_2024-02-22/ADMIN-EXPRESS-COG-CARTO/1_DONNEES_LIVRAISON_2024-03-00169/ADECOGC_3-2_SHP_LAMB93_FXX-ED2024-02-22"
folder_prod = "/Users/yann/Documents/Meio/R&D MILOO/PROD/geo/admin"
```

```python
# Import des régions et cleaning
reg = "REGION.shp"
reg = gpd.read_file(os.path.join(folder_raw, reg))
reg = reg.drop(columns=["ID","NOM_M"])
reg.rename(columns={"NOM":"label","INSEE_REG":'official_code'}, inplace=True)
reg['official_label'] = reg['label']
reg['class']="region"

reg.to_file(os.path.join(folder_prod,"regions_2015.shp"))
```

```python
com = "COMMUNE.shp"
com = gpd.read_file(os.path.join(folder_raw, com))
com = com.drop(columns=["ID","NOM_M"])
com.rename(columns={"NOM":"label","INSEE_COM":'official_code'}, inplace=True)
com['official_label'] = com['label']
com['class']="commune"

com.to_file(os.path.join(folder_prod,"communes_2024.shp"))
```

```python
def to_geo(gdf, geometry,  official_label, official_code,label,class_list,class_from_column=False ):
    """Fonction qui convertit un geodataframe en array de dictionnaires pour l'import dans une base de données MongoDB"""
    data = []
    geoj = reg.to_geo_dict()   
    for i in range(len(gdf)):
        doc = {
            'geometry': geoj['features'][0]["geometry"],
        }

        if label:
            doc['label'] = gdf[label].iloc[i]

        if not class_from_column:
            doc['class'] = class_list

        if class_from_column:
            class_list = [class_list] if isinstance(class_list, str) else class_list
            doc['class'] = gdf[class_from_column].iloc[i]
           
        if official_label:
            doc['official_label'] = gdf[official_label].iloc[i]
        
        if official_code:
            doc['official_code'] = gdf[official_code].iloc[i]
        
        data.append(doc)
    return data
```
